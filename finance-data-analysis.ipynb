#### Frequency of Age range by Gender 

```python
df.groupby('gender').age.plot(kind='kde')
plt.xlabel('age')
plt.legend()
plt.show()
```

#### Plotting the Gender of Respondents

```python
 sns.countplot(df['gender'],linewidth=3,palette="Set2",edgecolor='black')
plt.show()
```

From dataset male are more likely to invest than female

#### Plotting Type of Investment with Prefference

```python
df.iloc[:, 3:10].sum().plot(kind='bar')
plt.title('The sum of types of investments')
plt.ylabel('preference (less like)')
plt.show()
```

The rank in order of preferrence the data show people prefer to invest Public provident fund.

#### Plotting Age of Respondents

```python
sns.countplot(df['age'],palette="Set3",linewidth=2,edgecolor='black')
plt.show()
```

We can see from the above graph that most of our respondents were in the ages 24-31 and we will look further their habits in the finanace field 


#### Comparing factors of INVESTMENT_AVENUE & STOCK_MARKET with Gender

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer',linewidth=3,edgecolor='white')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='white')
plt.title('STOCK MARKET')
plt.show()
```

From the above plots we can conclude that Females mostly dont invest in the Stock Market

#### Comparing the Faactors that affect Investment

```python
 sns.countplot(x=df['Factor'],palette='coolwarm',linewidth=2,edgecolor='black')
```

The above trend shows us that Investors are less worried about risks involved and more worried about the Retruns on their Investments

#### Plotting Risk Factors by Gender

```python
sns.countplot(x=df['gender'],hue=df['Factor'],palette='Oranges',linewidth=2,edgecolor='black')
plt.title('Gender')
plt.show()
```

From the above plot we can see that higher number of Males are dependent on Returns

#### Plotting the Purpose of Investment by Age 

```python
plt.figure(figsize=(16,6))
sns.pointplot(x="Purpose",y='age',data=df,linestyles="--",capsize=.3,color='Red')
plt.show()
```

From the above plot it is clear that as people grow Older they are more inclined towards Returns and people at a Younger age are more inclined towards Wealth Creation

#### Plotting the Duration of Investemnt by Gender & Investment Monitoring cycles by Gender

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(hue=df['gender'],x=df['Duration'],palette='viridis',linewidth=2,edgecolor='black')
plt.title('DURATION')
plt.subplot(1, 2, 2)
sns.countplot(hue=df['gender'],x=df['Invest_Monitor'],palette='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()

```

Above graphs show us that most of the Investors Invest for a period of 1-5 Years and Monitor their Investments mostly on a Monthly basis

#### Finding out through which source Investors found about the opportunity the most

From the above plot we find out that Investors found out about the Investment oppotunity the most from Newspaper & Magazines
